# **Q.2 When people stay in hotels, how long do they usually stay?**

```python
# Let us find out night stay duration of customers
total_nights_stays = (df['stays_in_weekend_nights']+ df['stays_in_week_nights']).value_counts().head(10).sort_index()
print(total_nights_stays)
```

```python
# Plot the graph of number of nights people wants to stay against total number of bookings
total_nights_stays.plot.bar(color='blue')
plt.rcParams['figure.figsize'] = (13,7)
plt.grid()
plt.title('Night Stay Duration')
plt.ylabel('Total number of bookings')
plt.xlabel('Number of nights')
```

*   Generally, customers prefer to stay one, two, three, or four nights; however, bookings for two and three nights have a higher demand.

*   In order to get the best daily rate, we should consider staying for two to four days, since more people prefer the same length of stay. Because long-term stay offers will consume their rooms and new customers will not be able to take advantage of the rooms, hotels are also offering mid-term accommodation to generate extra profits.




```python
# Observing peoples interest to stay according to type of hotel with graph
df.loc[:,'total_nights'] = df['stays_in_weekend_nights']+ df['stays_in_week_nights']

fig, ax = plt.subplots(figsize=(12,6))
sns.countplot(x='total_nights', hue='hotel', data=df, order = df.total_nights.value_counts().iloc[:10].index)
ax.set_title('Hotel wise night stay duration')
ax.set_xlabel('No of Nights')
ax.set_ylabel('Count')
```


*   According to this visual representation, people are interested in city hotel to stay for two, three, one and four days respectively whereas for resort hotel people are interested to stay one, seven, two and three days respectively.

# **Q.5 What are the top 10 countries in terms of bookings?**

```python
top_10_countries = df[df['is_canceled']==0]['country'].value_counts()[:10]
top_10_countries = top_10_countries.reset_index().rename(columns = {'index':'country','country':'number_of_bookings'})
top_10_countries['percentage'] = (top_10_countries['number_of_bookings']/top_10_countries['number_of_bookings'].sum())*100
top_10_countries
```

```python
plt.figure(figsize=(10,10))
ax = sns.barplot(x="country", y="percentage", data=top_10_countries)
plt.title('Top 10 Countries by bookings percentage',fontweight="bold", size=20)
```


*   This graph show information about the countries that have most bookings. Most of bookings are created from Portugal and then Great Britain in second place and France at third place.


# **Q.6 Which was the most booked accommodation type (Single, Couple, Family)?**

```python
# Select single, couple, multiple adults and family
df_not_canceled = df[df['is_canceled'] == 0]

# Select single, couple, multiple adults and family
single   = df_not_canceled[(df_not_canceled.adults == 1) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
couple   = df_not_canceled[(df_not_canceled.adults == 2) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
family   = df_not_canceled[df_not_canceled.adults + df_not_canceled.children + df_not_canceled.babies > 2]
```

```python
# Make the list of Category names, and their total percentage
names = ['Single', 'Couple (No Children)', 'Family / Friends']
count = [single.shape[0],couple.shape[0], family.shape[0]]
count_percent = [x/df_not_canceled.shape[0]*100 for x in count]

# Creating a dictionary to make a DataFrame
d = {'names':names,'count_percent':count_percent}
acc_type = pd.DataFrame(data = d)
```

```python
# Showing the DataFrame
acc_type
```

```python
# Plotting the graph between acommodation type and their percentage of customer
plt.figure(figsize = (10,7), dpi=110)
sns.barplot(data = acc_type, x = 'names', y = 'count_percent')
plt.title('Accommodation Type', weight = 'bold')
plt.xlabel('Type of customer')
plt.ylabel('Booking (%)')
```


*   From graph, it is seen that most number of customer are containing 2 adults customer and those who have more than 2 either containing adults, children & babies have the lowest number of customer.
